```python
import numpy as np
import os

DATA_DIR = 'ravdess-emotional-song-audio'
SET_DIR = 'ravdess-emotional-song-spec'

def audio_to_spec(path, n):

    file_name = os.path.splitext(os.path.basename(path))[0]
    dir_name = os.path.dirname(path)
    class_name = dir_name.split("\\")[1]
    output_name = SET_DIR+'\\'+class_name+'\\'+class_name+'_'+str(n)+'.png'

    if not os.path.exists(SET_DIR+'/'+class_name):
        os.makedirs(SET_DIR+'/'+class_name)
    
    print(output_name)

    !"C:/Program Files (x86)/sox-14-4-2/sox.exe" $path -n spectrogram -X 56 -x 224 -y 224 -r -o $output_name

    return True

def dir_to_set(dir=None):
    for root, dirs, files in os.walk(dir, topdown=True):
        n = 0
        for name in files:
            audio_to_spec(os.path.join(root, name), n)
            n += 1

    return True


dir_to_set(DATA_DIR)
```
